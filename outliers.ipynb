```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create sample data
data = {
    'Group': ['A'] * 10 + ['B'] * 10,
    'Score': [50, 52, 49, 51, 53, 48, 52, 50, 300, 400,   # Group A with outliers (300, 400)
              60, 59, 58, 61, 62, 63, 64, 60, 58, 65]      # Group B (no outliers)
}

df = pd.DataFrame(data)

# Plot box plot
sns.boxplot(x='Group', y='Score', data=df)
plt.title("Box Plot with Outliers")
plt.show()
```
